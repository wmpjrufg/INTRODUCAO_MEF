Outra manipulação bastante relevante na fase de limpeza dos dados é em relação aos dados faltantes ( _missing values_ ). São diversas as possibilidades conforme citado no material base. Em um primeiro momento uma opção simplificada é a eliminação daquela linha de registro utilizando o método `.dropna()`.  Veja que a eliminação dos dados do tipo **NaN** ( vazio ou _null_ ) o _data frame_ é reduzido para um total 124.793,00 linhas. Isso configura uma redução de cerca 71,00% da base.

```python
base_carros_limpeza_2 = base_carros_limpeza_1.dropna()
print(base_carros_limpeza_2.shape)
base_carros_limpeza_2.head()
```

Outra maneira de visualizar o _dataset_ com os valores não preenchidos é com o uso do método `.isnull` que permitirá criar uma tabela de _booleanos_ contendo as informações **_True_** ou **_False_** a respeito de cada registro. Para visualização do _dataset_ anterior a limpeza basta consultar o objeto **base_carros_limpeza_1**.

```python
base_carros_dados_faltantes = base_carros_limpeza_1.isnull()
base_carros_dados_faltantes.head()
```

Para uma visualização detalhada da contribuição de dados faltantes ou dados preenchidos nesse _dataset_ é possível utilizar o comando. 

```python
for coluna in base_carros_dados_faltantes.columns.values.tolist():
    print("Coluna do dataframe = ", coluna)
    print(base_carros_dados_faltantes[coluna].value_counts())
    print(" ")
```

Porém em algumas ocasiões uma redução da base de dados dessa magnitude poderá afetar o resultado final da Mineração de Dados, logo seria interessante melhorar a base de dados com conhecimento de especialista ou técnicas similares que permitam completar os dados do tipo **NaN**.

São algumas possibilidades de substituição:
1. Substituição pela média, moda (em caso de valores categóricos);
2. Substituição pelo valor mais frequente;
3. Usar um outro modelo de Mineraçáo de Dados para determinar o valor;
4. Conhecimento de especialista.

Para exemplificar um sistema de substituição de dados de dados no _dataset_ vamos avaliar a kilometragem do veículo. A coluna correspondente a essa situação é o _id odometer_. São 75.148,00 dados faltantes nessa coluna. Portanto vamos considerar que a partir do ano de produção desse veículo a média anual de kilometragem foi de 15.000,00 kilometros. Portanto a equação que determinará a kilometragem será:

$novovalor = (dataatual - datafabric).15.000,00$

Portanto o valores deverão ser avaliados da seguinte forma:
- $novovalor$ corresponde ao novo valor de kilometragem que será colocado no _dataset_;
- $dataatual$ corresponde ao ano que analise da lista no caso estamos em 2020 logo $dataatual = 2020$;
- $datafabric$ corresponde ao id **year** do _dataset_
